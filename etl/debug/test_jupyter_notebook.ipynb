```python
#read a json file 
import json
#path = "/Users/sudheer.bomminenikroger.com/Downloads/subcommodity_dqa_prod_20241126.json"
path = "/Users/sudheer.bomminenikroger.com/Downloads/STAGE_SB.subcommodity_dqa.json"

#the json file is a list of dictionaries exported from a mongodb collection. Check if every json in each item of the list has non-empty keys 
with open(path) as f:
    data = json.load(f)

def check_json_keys(json_list):
    empty_key_items = []
    
    def check_nested_keys(obj):
        has_empty = False
        if isinstance(obj, dict):
            for key, value in obj.items():
                if key == "":
                    has_empty = True
                if isinstance(value, (dict, list)):
                    if check_nested_keys(value):
                        has_empty = True
        elif isinstance(obj, list):
            for item in obj:
                if check_nested_keys(item):
                    has_empty = True
        return has_empty

    for json_item in json_list:
        #print(f"Checking subCommodityCode: {json_item.get('subCommodityCode', 'No subCommodityCode found')}")
        if not json_item:
            print(f"Empty JSON object found")
            empty_key_items.append(json_item)
        if check_nested_keys(json_item):
            empty_key_items.append(json_item)
            
    return empty_key_items    

empty_key_items = check_json_keys(data)

#print all the empty key subcommoditycode
for item in empty_key_items:
    print(f"SubCommodityCode: {item.get('subCommodityCode', 'No subCommodityCode found')}")

#write the empty key items to a json file
import json
with open('/Users/sudheer.bomminenikroger.com/Downloads/empty_key_items.json', 'w') as f:
    json.dump(empty_key_items, f)
```
